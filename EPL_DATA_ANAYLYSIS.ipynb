```python
# CREATING NEW COLUMNS EXAMPLE IS CHECKING THE MINUTE PER MATCH OF EVERY PLAYER AND GOALS PER MATCH
data['minsPerMatch'] = data['Mins'] / data['Matches']

data['goalsPerMatch'] = data['Goals'] / data['Matches']

data.head()
```

```python
# CHECKING THE TOTAL NUMBER OF PENALTY GOAS SCORED
penanlty_goals = data['Penalty_Goals'].sum()

all_goals = data['Goals'].sum()

# TO CHECK THE NON PENALTY GOALS 
# SUBTRACT THE PENALTY GOALS FROM THE GOALS COLUMN
non_penalty_goals = all_goals - penanlty_goals

non_penalty_goals
```

```python
# PIE CHART SHOWING THE PENALTY SCORED AND PENALTY MISSED
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.head()
```

```python
# PLOTTING ON A PIE CHART THE NUMBER OF PENALTY SCORED AND THE NUMBER OF PENALTY MISSED

plt.figure(figsize = (13,6))
penalty_not_scored = data['Penalty_Attempted'].sum() - data['Penalty_Goals'].sum()
data = [penalty_not_scored,data['Penalty_Goals'].sum()]
explode = (0, 0.1) 
labels = ['penalty missed','penalty scored']
color = sns.color_palette('Set2')
plt.pie(data,labels = labels,colors = color, autopct = '%1.1f%%',explode=explode)
plt.show()
```
